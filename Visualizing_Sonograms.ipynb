# Sonogram Visualization and some Initial Statistics

This data comes from Richard Mooney's group and are audio recordings of a zebra finch bird singing. The data are organized from day 60 to day 90, where the day indicates the age of the bird.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

I have already precomputed sonograms for each audio recording. I want to load some sonograms from day 61 and visualize them.

```python
folder_path = '/home/akapoor/Desktop/Juvenile_ZF/Python_Objects/61'
files = os.listdir(folder_path)
dat = np.load(f'{folder_path}/{files[0]}')
```

```python
# Assuming 'datasets' is a list of 10 datasets (for example, images)
datasets = [np.load(f'{folder_path}/{files[i]}') for i in range(10)]  # Replace 'your_data' with your actual datasets
```

```python
# Create a figure and a grid of subplots
fig, axs = plt.subplots(2, 5, figsize=(20, 8))

# Iterate over each subplot to display each dataset
for i in range(2):  # Loop over the rows
    for j in range(5):  # Loop over the columns
        index = i * 5 + j  # Calculate the index of the dataset
        dat = datasets[index]
        axs[i, j].pcolormesh(dat['time'], dat['freq'], dat['spec'], cmap = 'inferno')
        axs[i, j].set_title(f'Sonogram {index}')  # Set title for each subplot
        axs[i, j].set_xlabel("Time (s)")
        axs[i, j].set_ylabel("Frequency (Hz)")
        # axs[i, j].axis('off')  # Optional: Remove axes for cleaner look

plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()
```

We can see that there is a lot of variability in the sonograms.
